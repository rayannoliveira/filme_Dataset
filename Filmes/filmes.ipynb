## Importando o banco de dados 

```python
df = pd.read_csv('tmdb_5000_movies.csv')
df.head(5)
```

 ## 1- Qual país produz mais filme ?

```python
df.original_language.value_counts()
```

```python
lingua_ori = pd.DataFrame(df.original_language.value_counts()) 
```

```python
lingua_ori
```

```python
lingua_ori.reset_index(inplace = True)
lingua_ori
```

```python
lingua_ori.columns = ['original_language','total']
```

```python
import seaborn as sns
```

```python
sns.histplot(data=lingua_ori, x="original_language")
```

```python
sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
ax = sns.barplot(x="original_language", y="total", data=lingua_ori)
```

```python
# Qual o pais com menos filme ?
crescente_lingua_ori= pd.DataFrame()
crescente_lingua_ori= lingua_ori
crescente_lingua_ori
```

```python
crescente_lingua_ori.sort_values(by=['total'],
                         ignore_index=False,
                                 inplace= True,
                         ascending=True)
```

```python
crescente_lingua_ori
```

```python
sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
ax = sns.barplot(x="original_language", y="total", data=crescente_lingua_ori)
```

## 3- Qual teve maior numero de votos ?

```python
#Verificando qual o maior numero
df.vote_count.max()
```

```python
# colocando em ordem o maior numero de notas em ordem decrescente 
df.sort_values('vote_count', ascending = False)
```

```python
# Separando em um novo DataFrame com somente os valores de vote_count e original_title 
n_votos_titulo = pd.DataFrame(df[['vote_count','original_title']])
n_votos_titulo.sort_values('vote_count', ascending = False)
```

## 4- Qual filme teve mais votos com menos valor de orçamento maximo?  

```python
# Necessário fazer algoritmo 
len(df)
```

```python
df[df['budget'] >= 380000000]
```

```python
df.budget.describe()
```

```python
df.vote_count.describe()
```

```python
df[df['vote_count'] <= 13752]
```

```python
## Comparando variaveis 
selecao = (df['budget'] < 380000000) & (df['vote_count'] > 4948)
mais_voto_menos_ocamento= df[selecao]
mais_voto_menos_ocamento.sort_values('vote_count', ascending = False)
```
